```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# Path to the NetCDF file
file_path_1 = '/Users/calebburke/Downloads/PyFT/PS_Data/ORIG2_star01-Kepler-SolarLike_PS.nc'
file_path_2 = '/Users/calebburke/Downloads/PyFT/PS_Data/star02-Kepler-SolarLike_PS.nc'
file_path_2 = '/Users/calebburke/Downloads/PyFT/PS_Data/star03-Kepler-DeltaScuti_PS.nc'


# Open the NetCDF file using xarray
ds1 = xr.open_dataset(file_path_1)
ds2 = xr.open_dataset(file_path_2)
ds3 = xr.open_dataset(file_path_2)

ds1['freq'].attrs['units'] = 'hr⁻¹'
ds1['P'].attrs['units'] = 'ppm²'
plt.figure(figsize=(16, 6))
ds1['P'].sel(freq=slice(5,10)).plot(color='black', alpha=0.5, label='Low Freq Res Data')
ds1['P'].sel(freq=slice(7,7.06)).plot(color='darkorange', label='Peak 1: $\ell = 1$')
ds1['P'].sel(freq=slice(7.75,7.8)).plot(color='darkgreen', label='Peak 2: $\ell = 1$')
ds1['P'].sel(freq=slice(8.3,8.36)).plot(color='red', label='Peak 3: $\ell = 0$ & $2$')

#plt.ylim(0,10)
plt.title('Power Spectrum of Star01-Kepler-SolarLike_PS')
plt.legend()
plt.savefig('/Users/calebburke/Downloads/PyFT/plots/LongSpectrum.png')
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt

# Load datasets
peak_3 = '/Users/calebburke/Downloads/PyFT/PS_Data/ZOOM3_star01-Kepler-SolarLike_PS.nc'
ds_p3 = xr.open_dataset(peak_3)

peak_2 = '/Users/calebburke/Downloads/PyFT/PS_Data/ZOOM2_star01-Kepler-SolarLike_PS.nc'
ds_p2 = xr.open_dataset(peak_2)

peak_4 = '/Users/calebburke/Downloads/PyFT/PS_Data/ZOOM4_star01-Kepler-SolarLike_PS.nc'
ds_p4 = xr.open_dataset(peak_4)

# Create a figure with 3 subplots in a single row
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot the first dataset
ds_p3['P'].plot(ax=axes[0], color='orange', alpha=0.8, label='High Freq Res')
ds1['P'].sel(freq=slice(7, 7.06)).plot(ax=axes[0], color='black', linestyle='--', label='Low Freq Res')
axes[0].legend()
axes[0].set_title('Peak 1')

# Plot the second dataset
ds_p2['P'].plot(ax=axes[1], color='green', alpha=0.8, label='High Freq Res')
ds1['P'].sel(freq=slice(7.75, 7.8)).plot(ax=axes[1], color='black', linestyle='--', label='Low Freq Res')
axes[1].legend()
axes[1].set_title('Peak 2')

# Plot the third dataset
ds_p4['P'].plot(ax=axes[2], color='red', alpha=0.8, label='High Freq Res')
ds1['P'].sel(freq=slice(8.3, 8.36)).plot(ax=axes[2], color='black', linestyle='--', label='Low Freq Res')
axes[2].legend()
axes[2].set_title('Peak 3')

plt.tight_layout()
plt.savefig('/Users/calebburke/Downloads/PyFT/plots/ZoomedModes.png')
plt.show()
```
